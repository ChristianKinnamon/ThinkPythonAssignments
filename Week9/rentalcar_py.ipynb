```python
#Name: rentalcar.py

#Import Statements
#(None)

# FUNCTION DEFINITIONS

# 1) Define get_classification() get and validate B/S/L code

# 2) Define calculate_distance() which will compute total miles driven

# 3) Define calculate_bill() will calculate total bill based on days and mileage

# 4) Define get_classification_name() converts code to full description

#PROGRAM LOGIC

# 1) Prompt user for classification code appears

# 2) Prompts for start and end odometer appears

# 3) number of rental days prompt appears

# 4) calculate total miles

# 5) calculate total bill and charged miles

# 6) Get full classification name

# 7) Display summary

def validate_classification(code):
    return code == 'B' or code == 'S' or code == 'L'

def class_name(code):
    if code == 'B': return "Budget"
    if code == 'S': return "Standard"
    if code == 'L': return "Luxury"

def main():
    code = input("Enter classification code (B,S,L): ").upper()
    while not validate_classification(code):
        code = input("Invalid. Enter B, S, or L: ").upper()

    start = int(input("Start Odometer: "))
    end = int(input("End Odometer: "))
    days = int(input("Days Rented: "))

    miles = end - start

    # For commit 1, just show input and distance
    print("\n--Rental Summary--")
    print(f"Classification code: {code}")
    print(f"Total Miles: {miles}")
    print(f"Days Rented: {days}")

if __name__ == "__main__":
    main()
```
