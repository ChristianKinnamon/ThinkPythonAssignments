```python
# Import Statements

# (None)


# Function Definitions

# 1) Define Book class

# 2) Define __init__(self, title, author)
# Assign title and author to the object
# Set is_available to True by default

# 3) Define __str__(self)
# Return formatted string with title and author
# Show as "Available" if is_available is True
# Show as "Checked Out" if is_available is False

# 4) Define checkout(self)
# Set is_available to False

# 5) Define return_book(self)
# Set is_available to True


# Program Logic

# 1) Create two Book objects

# 2) Print both of the books

# 3) Call checkout() on one book

# 4) Print both books again to show updated status

class Book:
  def __init__(self, title, author):
    self.title = title
    self.author = author
    self.is_available = True

  def __str__(self):
        status = "Available" if self.is_available else "Checked Out"
        return f"{self.title} by {self.author} - {status}"

  def checkout(self):
      self.is_available = False

  def return_book(self):
      self.is_available = True

book1 = Book("The Boy Who Harnessed the Wind", "William Kamkwamba")
book2 = Book("And Then There Were None", "Agatha Christie")

print("-- Library Status --")
print(book1)
print(book2)
print()
print("Checking out 'And Then There Were None'...")
book2.checkout()
print("-- Updated Status --")
print(book1)
print(book2)
```
