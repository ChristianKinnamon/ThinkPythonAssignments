```python
# IMPORT STATEMENTS

# (None)


# FUNCTION DEFINITIONS

# 1 Define write_entry(filename,entry)

# Open file in append mode ('a')

# Write the entry to the file

# Add a newline character (\n)

# Close the file


# 2 Define read_journal(filename)

# Open the file in read mode ('r')

# Read contents of file

# Print contents of file

# Close file


# Program Logic

# Ask the user if they want to (W)rite a new entry or (R)ead the journal

# If they choose (W), ask for their thoughts and call write_entry

# If they chose (R), call read_journal

# Wrap this in a while loop so the user can keep using the app until they type "quit"



def write_entry(filename, entry):
    file = open(filename, 'a')
    file.write(entry + "\n")
    file.close()

def read_journal(filename):
   try:
       file = open(filename, "r")
       contents = file.read()
       print(contents)
       file.close()
   except FileNotFoundError:
       print("No Journal Entries Found Yet.")


filename = "journal.txt"

while True:
     choice = input("(W)rite, (R)ead, or (Q)uit? ")

     if choice == "W":
       entry = input("Enter your entry: ")
       write_entry(filename, entry)
       print("Entry Saved.")

     elif choice == "R":
       print("--- Your Journal ---")
       read_journal(filename)

     elif choice == "Q":
       break

     else:
       print("Invalid choice.")
```
