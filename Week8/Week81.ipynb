```python
#Import Statements

# (none)

#Function Definitions
# create function is_palidrome (word)
# make the word case insensitive
# reverse word through string slicing
# compare values, return True or False

#Program Logic
# get user input for the word
# call the function and print the words "word" is a palidrome if true
# if false then print the words "words" isn't a palidrome

def is_palidrome(word):
  word = word.lower()
  reverse = word[::-1]
  if reverse == word:
     return True
  else:
     return False

word = "Racecar"
is_palidrome(word)
```
