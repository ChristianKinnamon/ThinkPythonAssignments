```python
# NAME: roster.py

#Import Statements

# (None)

# Function Def

# 1) define add_players(roster)

# Use a While loop to let the player enter names

# Stop the loop when the user types quit

#  All names entered (except quit) should be added to the list

# 2) Define is_on_roster

# Returns True if the player is on the list

# Returns False if not

# Program Logic

# 1) create a list called "roster"

# 2) Call add_players to fil the list

# 3) Ask user for a specific name

# 4) Call is_on_roster

# 5) use the return value to print a message

def add_players(roster):
    while True:
          name = input("Enter player name(or 'quit' to stop): ")

          if name == "quit":
             break

          roster.append(name)

roster =[]

add_players(roster)

print(roster)
```
