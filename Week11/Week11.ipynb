```python
# Import data

# (none)


# Function Definitions

# 1) define add_students(gradebook):

# if "done" is entered exit loop

# otherwise ask for students grade 0-100

# store students name and use as a key with the grade being the value

# 2) define calculate_average(gradebook):

# be able to access grade values from gradbook

# calculate all student grades

# return the calculated number


# Program Logic

# 1) initialize an empty dictionary called gradebook.

# 2) Call add_statement to populate dictionary.

# 3) Call calculate_average and display the resulting class average

# 4) Print a final report showing each student's name and their grade



def add_students(gradebook):

    while True:
          student_name = input("Enter student name (or 'done'): ")
          if student_name.lower() == "done":
            break
          grade = int(input("Enter {student_name}'s grade: "))
          gradebook[student_name] = grade

def calcualte_average(gradebook):
    if not gradebook:
      return 0.0
    total = sum(gradebook.values())
    count = len(gradebook)
    average = total / count
    return average

gradebook= {}
add_students(gradebook)

average = calcualte_average(gradebook)

print("---Class Report---")
for student, grade in gradebook.items():
    print(student + ": " + str(grade))
print("Class Average: " + "{:.2f}".format(average))



```
